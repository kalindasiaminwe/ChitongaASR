This notebook splits the session audio recordings into the dev.csv, test.csv and train.csv

```python
import pandas as pd

dataset = pd.read_csv('/content/Sample-Spreadsheet-100-rows.csv')


from sklearn.model_selection import train_test_split
data_main, test = train_test_split(dataset, test_size=.2, random_state=23)
train, validate = train_test_split(data_main, train_size=.75, random_state=23)

train.to_csv('./train.csv', sep='\t',index=False)
test.to_csv('./test.csv', sep='\t',index=False)
validate.to_csv('./validate.csv', sep='\t',index=False)
```

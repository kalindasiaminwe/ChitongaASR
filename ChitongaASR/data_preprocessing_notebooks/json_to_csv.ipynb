This notebook creates a csv file containing the following: 
1.   audio id
2.   duration
3.   bits per sample
4.   gender
5.   lan of recording
6.   mother tongue


```python
import pandas as pd
import json
import os
import glob

#Function to extract files from folder
def extract_files(path):
    all_files = []
    for root, dirs, files in os.walk(path):
        files = glob.glob(os.path.join(root,'*.json'))
        for f in files :
            all_files.append(os.path.abspath(f))
    return all_files


info = extract_files('C:/Users/Kalinda Siaminwe/Documents/4th year/Final year project/notebooks/json_files')
```

```python
#check if the files are accessed successfully
filepath = info[0]

df = pd.read_json(filepath, lines=True)
df.head()
```

```python
#Create a list of all the json files
json_list = []

for file in info:
    with open(file) as f:
        for line in f:
            json_list.append(json.loads(line))
```

```python
json_list[:2]
```

```python
json_files = pd.DataFrame(json_list)
json_files.head()
```

```python
#make a copy of the dataframe
df = json_files.copy()

#drop the columns that are not needed
df.drop(['date', 'origin', 'source', 'suffix', 'Format', 'file_type', 'speakers', 'speaker_note', 'item_id', 'speaker_birth_year','speaker_name', 'languages', 'version', 'sampleRate', 'user_id', 'NumChannels', 'location', 'device', 'androidID'], axis=1, inplace=True)
```

```python
#check the dataframe
df.head()

```

```python
#rename the column name to audio_id
df.rename(columns = {'name':'audio_id'}, inplace = True)
df.head()
```

```python
#rearange the columns
df = df[['audio_id', 'recording_lang', 'BitsPerSample','durationMsec', 'speaker_gender', 'mother_tongue']]
df.head()
```

```python
#save the dataframe as a csv file
df.to_csv('recordings_info.csv', index=False)
```
